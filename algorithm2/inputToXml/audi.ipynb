```python
import subprocess
import os
```

```python
def convert_image_to_musicxml(image_path, output_dir):    
    audiveris_path = "/Users/tao-taohe/audiveris/audiveris/build/distributions/Audiveris-5.3.1/bin"
    # Create the command with the output directory specified
    command = [
        os.path.join(audiveris_path, 'audiveris'),
        '-export', 
        image_path,
        '-output', 
        output_dir
    ]

    try:
        # Run the command
        subprocess.run(command, check=True)
        print("Audiveris has successfully processed the file.")
        
        # Check for the output file
        output_file = os.path.join(output_dir, os.path.basename(image_path).replace('.png', '.musicxml'))
        if os.path.exists(output_file):
            print(f"Output file created: {output_file}")
        else:
            print("Output file not found.")
            
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")
```
